```
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio
```

```
def draw_boxes(image,bounds,color='yellow',width=2):
  draw = ImageDraw.Draw(image)
  for bound in bounds:
    p0,p1,p2,p3 = bound[0]
    draw.line([*p0,*p1,*p2,*p3,*p0],fill=color,width=width)
  return image
```

```
draw_boxes(im,bounds)
```

```
text_list = reader.readtext('/content/images?q=tbn:ANd9GcSXvUFEduM5QpY1fvMOC4LC8MYUAYfE8XeR_Q', add_margin=0.55, width_ths=0.7, link_threshold=0.8, decoder='beamsearch',blocklist='=-', detail=0)


text_list
```

```
text_combo =" ".join(text_list)
```

```
text_combo
```

## Use google translator to convert the text into english

```
translator = Translator()
print(translator.detect(text_combo))
```

```
text_en = translator.translate(text_combo,src='hi')
print(text_en.text)
```

```
hi_tts =gTTS(text_en.text)
hi_tts.save('trans.mp3')
```

```
Audio('/content/trans.mp3',autoplay=True)
```

## Play in indian ascent

```
hi_tts =gTTS(text_en.text,lang='hi')
hi_tts.save('transhindi.mp3')
Audio('/content/transhindi.mp3',autoplay=True)
```

# Play the audio in french ascent

```
hi_tts =gTTS(text_en.text,lang='fr')
hi_tts.save('transhindi.mp3')
Audio('/content/transhindi.mp3',autoplay=True)
```

# Convert text to regional language - odiya

```
text_en = translator.translate(text_combo,src='hi',dest='or')
print(text_en.text)
```

```
hi_tts =gTTS(text_en.text,lang='gu')
hi_tts.save('gujrati.mp3')
Audio('/content/gujrati.mp3',autoplay=True)
```
